```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels
import statsmodels.api as sm
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
advertising = pd.read_csv("advertising.csv")
```

```python
advertising.head()
```

```python
advertising.shape
```

```python
advertising.info()
```

```python
advertising.describe()
```

```python
for param in ["TV", "Radio", "Newspaper"]:
    sns.regplot(x=param, y="Sales", data=advertising)
    plt.title(f"{param} vs Sales")
    plt.show()
```

```python
sns.pairplot(data=advertising, x_vars=["TV", "Radio", "Newspaper"], y_vars="Sales")
plt.show()
```

```python
sns.heatmap(advertising.corr(), annot=True, cmap="Greens")
plt.show()
```

```python
X = advertising["TV"]
y = advertising["Sales"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=100)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train_sm = sm.add_constant(X_train)
```

```python
X_train_sm.head()
```

```python
lr = sm.OLS(y_train, X_train_sm)
lr_model = lr.fit()
lr_model.params
```

```python
lr_model.summary()
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, 6.9487 + 0.0545 * X_train, "red")
plt.show()
```

```python
y_train_predict = lr_model.predict(X_train_sm)
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, y_train_predict, "red")
plt.show()
```

```python
res = y_train - y_train_predict
```

```python
fig = plt.figure()
sns.displot(res, kde=True, bins=15)
fig.suptitle('Error Terms', fontsize = 15)                  # Plot heading 
plt.xlabel('y_train - y_train_pred', fontsize = 15)         # X-label
plt.show()
```

```python
sns.kdeplot(res)
plt.show()
```

```python
plt.scatter(X_train, res)
plt.show()
```

```python
X_test_sm = sm.add_constant(X_test)
y_test_pred = lr_model.predict(X_test_sm)
```

```python
r2 = r2_score(y_true=y_test, y_pred=y_test_pred)
```

```python
r2
```

```python
mean_squared_error(y_true=y_test, y_pred=y_test_pred)
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_test_pred, "red")
plt.show()
```
