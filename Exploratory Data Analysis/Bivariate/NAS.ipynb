```python
pivot_table = df.pivot(columns=['Age','Father.edu'], values='Science..')
```

```python
plt.figure(figsize=(50,60))
sns.heatmap(pivot_table, annot=True, fmt="g", cmap='Blues', cbar_kws={'label': 'Science..'})
plt.title('Heatmap of Values by Category1 and Category2')
plt.xlabel('Category2')
plt.ylabel('Category1')
```

```python
age = df.Age.str.contains("11")
df[age]
```

```python
df_filtered = df[df['Age'] != 11]

# Create a pivot table for the heatmap
pivot_table = df_filtered.pivot_table(index='Age', columns='Father.edu', values='Science..', aggfunc='mean')

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, cmap='viridis', cbar_kws={'label': 'Average Science Marks'})
plt.title('Heatmap of Average Science Marks by Age and Father\'s Education Level')
plt.xlabel('Father\'s Education Level')
plt.ylabel('Age')
plt.show()
```
