```python
import numpy as np
import matplotlib.pyplot as plt

months = np.array(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
sales = np.array([241268.56, 184837.36, 263100.77, 242771.86, 288401.05, 401814.06, 258705.68, 456619.94, 481157.24, 422766.63, 555279.03, 503143.69])

plt.plot(months, sales)

plt.title("Sales in 2015", fontdict={"fontsize": 20, "color": "green", "fontweight": 10 })
plt.xlabel("Months", fontdict={"fontsize": 12, "color": "red", "fontweight": 10 })
plt.ylabel("Sales", fontdict={"fontsize": 12, "color": "red", "fontweight": 10 })

ticks = np.arange(0, 600000, 50000)
label = ["{}K".format(i//50000) for i in ticks]

plt.yticks(ticks, label)
plt.xticks(rotation=45)

for xy in zip(months, sales):
    plt.annotate("{}K".format(xy[1]//1000), xy = xy, textcoords="data")

plt.show()
```
